# Grid Search to find optimal parameters

* C, (10<sup>-3</sup> to 10<sup>3</sup>)
* penalty (l1, l2) 
* multi_class (ovr, multinomial)

```
mnist = fetch_openml('mnist_784', version=1)
X, y = mnist["data"], mnist["target"]
X = X/255.0
X_scaled = scale(X)
print(X.shape)
print(y.shape)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.3, train_size = 0.7, random_state = 10)
print(X_train.shape)
print(X_test.shape)
```

```
model = LogisticRegression(solver='lbfgs', C=1, multi_class='multinomial', max_iter=2000)
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```

```
print("Accuracy: ", metrics.accuracy_score(y_test, predictions))
```
