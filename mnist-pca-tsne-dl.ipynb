```
from sklearn.datasets import fetch_openml
import numpy as np
from sklearn import svm, metrics
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns

%reload_ext autoreload
%autoreload 2
%matplotlib inline

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```
mnist = fetch_openml('mnist_784', version=1)
X, y = mnist["data"], mnist["target"]
X = X/255.0
X_scaled = scale(X)
print(X.shape)
print(y.shape)

# X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.3, train_size = 0.1, random_state = 10)
# print(X_train.shape)
# print(X_test.shape)
```

```
model = DecisionTreeClassifier(max_depth=10, max_leaf_nodes=100)
X_PCA = PCA(n_components=40).fit_transform(X)
X_embedded = TSNE(n_components=2).fit_transform(X_PCA)
```

```
X_train, X_test, y_train, y_test = train_test_split(X_embedded, y, test_size = 0.3, train_size = 0.7, random_state = 10)
```

```
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```

```
print("Accuracy: ", metrics.accuracy_score(y_test, predictions))
```
